```python
#線形代数用ライブラリ
import scipy.linalg as linalg

#最小化計算（最小）用の関数
from scipy.optimize import minimize_scalar
```

```python
#行列計算
matrix = np.array([[1,-1,-1], [-1,1,-1], [-1,-1,1]])

print('行列式：', linalg.det(matrix))
print('逆行列：')
print(linalg.inv(matrix))
print(matrix.dot(linalg.inv(matrix)))
print()

#固有値と固有ベクトル
eig_value, eig_vector = linalg.eig(matrix)
print('固有値：', eig_value)
print('固有ベクトル：', eig_vector)
```

```python
#ニュートン法
#テキトーな方程式の解を求める
def my_function(x):
    return (x**2 + 2*x +1)

#ニュートン法の読み込み
from scipy.optimize import newton

#計算
print(newton(my_function, 0)) #f(x)=0の解を求める
```

```python
#単峰関数の極値を求める
print(minimize_scalar(my_function, method='Brent'))
```

```python
#practice 2
x1 = np.random.uniform(0,1,10000)
x2 = np.random.uniform(0,1,10000)

dis = []
in_circle = []
out_circle = []
cnt = 0
for a,b in zip(x1, x2):
    dis_tmp = np.sqrt(a**2 + b**2)
    dis.append(dis_tmp)
    if dis_tmp <= 1:
        cnt += 1
        in_circle.append([a,b])
    else:
        out_circle.append([a,b])
print(cnt)

in_circle = np.array(in_circle)
out_circle = np.array(out_circle)
plt.figure(figsize=(10, 10))
plt.scatter(in_circle[:,0], in_circle[:,1])
plt.scatter(out_circle[:,0], out_circle[:,1])
plt.show()

print('円周率',len(in_circle)*4/(len(in_circle)+len(out_circle)))
```
